```python
import requests
import sqlalchemy
import pandas as pd
from pandas import json_normalize 
from sqlalchemy.orm import sessionmaker
import json
from datetime import datetime
import datetime
import sqlite3
import os

# CONSTANTS

CLIENT_ID = 'CLIENT_ID'
CLIENT_SECRET = 'CLIENT_SECRET'
TOKEN = 'TOKEN'

# Request

if __name__ == "__main__":
    headers = {
        "Authorization":"Bearer {token}".format(token=TOKEN),
    }

today = datetime.datetime.now()
yesterday = today - datetime.timedelta(days = 1)
yesterday_unix_timestamp = int(yesterday.timestamp()) * 1000
time = yesterday_unix_timestamp


# Recently Played, MAX 50
r = requests.get(
    #'https://api.spotify.com/v1/me/player/recently-played?limit=50&after={time}'.format(time=time), headers=headers)
    'https://api.spotify.com/v1/me/player/recently-played?limit=50', headers=headers)

# Flattening
data = r.json()
df = json_normalize(data, 'items')

# Filtering Data down to wanted columns.
df = df.filter(
    [
    'played_at',
    'track.album.album_type', 
    'track.album.artists',
    'track.album.available_markets',
    'track.album.id',
    'track.album.name', 
    'track.album.release_date',
    'track.album.release_date_precision',
    'track.album.total_tracks',
    'track.album.type',
    'track.artists',
    'track.available_markets',
    'track.disc_number',
    'track.duration_ms',
    'track.explicit', 
    'track.id',
    'track.is_local', 
    'track.name', 
    'track.popularity',
    'track.track_number', 
    'track.type'
    ]
)

# Creating DataFrame
dataset = pd.DataFrame(df)
dataset

# Writing DataFrame to CSV.
df.to_csv(r'C:\Users\Trame\Documents\Fullstack Academy\Projects\Spotify Project\API Data\SpotifyAPIGet.csv', mode='a+',index=False, header=False)
```
