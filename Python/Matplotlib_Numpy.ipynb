```python
import matplotlib.pyplot as plt
import numpy as np

# xpoints = np.array([2,2,6,15])
ypoints = np.array([3,8,1,10])

# ms = marker shape
# mfc = marker face color

plt.plot(ypoints,'o:g', ms = 20, mfc = 'g')
plt.title('Hello There Friend')
plt.xlabel('Bing Bong')
plt.ylabel('Oh Wow')
plt.show()
```

```python
x = np.array(['A','B','C','D'])
y = np.array([5, 10, 15, 20])

plt.bar(x,y, color =['red','blue','purple','green'])

plt.show()
```

```python
y = np.array([25,35,15,25])
mylabels = ['dogs','cats','birds','fish']
mycolors = ['green','blue','yellow','black']
plt.pie(y, labels = mylabels, colors = mycolors)

plt.show()
```

```python
x = np.array([0,2,3,4])
y = np.array([5,8,10,11])

plt.subplot(1,2,1)
plt.title('First Plot Data')
plt.plot(x,y)

x = np.array([0,2,4,4])
y = np.array([3,10,2,9])

plt.subplot(1,2,1)
plt.title('Second Plot Data')
plt.plot(x,y)


plt.suptitle('Both Charts')



plt.show()
```

```python
x = np.array([0,2,3,4])
y = np.array([5,8,10,11])
colors = np.array([0,10,20,30])

plt.scatter(x, y, c = colors, cmap='rainbow')
plt.colorbar()
plt.show()
```
