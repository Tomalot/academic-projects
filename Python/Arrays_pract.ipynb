```python
xpts = np.array([1, 2, 3, 4, 5, 6])
ypts = np.array([4, 9, 1, 11, 6, 8])

plt.plot(xpts, ypts)

plt.show()
```

```python
# Arrays need to have == element counts

xpts = np.array([5,6,7,8,9,10,11,19,33,5])
ypts = np.array([23,65,23,67,12,89,46,34,9,10])

plt.scatter(xpts, ypts)

plt.show()
```
